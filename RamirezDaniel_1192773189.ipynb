```python
#EJERCICIO 7:
A = int(input('Digite el numero donde comienza la lista:'))
B = int(input('Digite un numero mayor donde termina la lista:'))
LISTA = list(range(A,B))

for i in range(B-A):
    if LISTA[0]<LISTA[0+i]:#Verifica que la lista este organizada comparando
                           #el primer el indice 0 con los posteriores
        print(f'La lista {LISTA} esta organizada de forma ascendente')
        break
C = int(input('''Digite un numero entre los dados anteriormente para ingresar
a las lista:'''))

for i in range(B-A):
    if LISTA[i] > C: #Itera en la LISTA y cuando encuentre el numero mayor a C
                     #li insertara en la lista
        LISTA.insert(i,C)
        print('La nueva lista es:',LISTA)
        break
```

```python
#EJERCICIO 8:
A=int(input('Digite un primer elemento de la lista:'))
B=int(input('Digite el segundo elemento de la lista:'))
C=int(input('Digite el tercer elemento de la lista:'))
D=int(input('Digite el cuarto elemento de la lista:'))
E=int(input('Digite el quinto elemento de la lista:'))
lista = [A,B,C,D,E]
X =(lista[0]+lista[1]+lista[2]+lista[3])//4 #Promedia los elementos de la lista
print('La lista es:',lista)


MAYOR = 0
SEGMAYOR = 0
for i in lista: #i tomara el valor del numero mayor en la lista y se eliminara
                #de dicha lista
    if i > MAYOR:
        MAYOR = i   
lista.remove(MAYOR)

while X==1:     #Si todos los numeros de la lista son iguales el promedio es 1
                #entonces no hay numero mayor
    print('No hay numero mayor')
    break   
while X!=1:     #Si el promedio es diferente de 1 los elementos de la lista no
                #son iguales entre si, entonces i tomara el valor del mayor 
                #numero de la nueva lista
    for i in lista:
        if i > SEGMAYOR:
            SEGMAYOR = i 
    
    print('El segundo numero mayor es:',SEGMAYOR)
    break
```

```python
#EJERCICIO 10:
from random import randint #se importa la libreria para generar numeros
                           #aleatorios
A = randint(0,99)
lista1 = []           
lista2 = []
lista3 = []
lista4 = []
matriz = [[lista1],
          [lista2],   
          [lista3],                 
          [lista4]]
B = 0                        #Esta parte del procedimiento es para generar una
                             #matriz cuyos elementos sean aleatorios
while B < 5:                       
    lista1.append(randint(0,99))
    lista2.append(randint(0,99))
    lista3.append(randint(0,99))
    lista4.append(randint(0,99))
    B += 1
print('la matriz es:',matriz)
AA = lista4[::-1]
BB = lista3[5:0:-1]   #Se utiliza indexado para organizar los elementos de la 
                      #matriz en una lista de la forma deseeada
vector=[AA[:],lista3[0],lista2[0],lista1[0:5],lista2[4],
        BB[:],lista2[1:4]]
VECTOR = []
for i in vector: #Como hay una lista anidadas se utiliza el ciclo for para 
                 #''aplanarla'' por asi decirlo
    if type(i) == int:
        VECTOR.append(i)
    elif type(i) == list:
        for e in i:
            VECTOR.append(e)
print('El vector solicitado es:',VECTOR)
```

```python
#EJERCICIO 11:
ancho=int(input('Digite la anchura del rectangulo:'))
alto=int(input('Digite la altura del rectangulo:'))
caracter=input('Digite un caracter:')

def fun():
    """Utilizando a c como contador que empieza en cero, cada vez se le va a
    ir agregando +1 hasta que iguale a alto, mientras imprime el producto del
    caracter por el ancho"""
    c = 0
    while c<alto:
        c+=1
        print(caracter*ancho)
fun()
```

```python
#EJERCICIO 12:
a = int(input('Anchura del triangulo:'))

def fun():
    """ Usando como contador i que empienza en 0, mientras i sea menos
    a 0, se le sumara 1 a la vez que se imprime el producto del asterico
    las veces que se repita el ciclo"""
    i = 0
    while i<a:
        i+= 1
        print ('*'*i)
    
    while i>0:
        """ Usando como contador i que empienza en a-1, mientras i sea mayor
        a 0, se le restara 1 a la vez que se imprime el producto del asterico
        las veces que se repita el ciclo"""
        i-= 1
        print ('*'*i)
fun()
```
